Istotą modelowania probabilistycznego jest opisanie  wszystkich nieznanych aspektów systemu modelu jako zmienne losowe a następnie wyznaczanie parametrów tych zmiennych.
Typowym przykładem może być dopasowywanie krzywej do punktów. W tym przypadku zakładamy znaną postać zależności, ale nie znamy błędów , dlatego błędy reprezentujemy jako zmienne losowe o średniej 0. Proces dopasowywania do estymacja parametrów tych zmiennych.
Bardzo często losowość jest tak duża, że nie możemy wskazać bezpośredniej zależności do dopasowanie, lub interesuje nas sam rozkład. W takiej sytuacji mieszanka rozkładów potrafi się dopasować do bardzo skomplikowanych danych.

### **Zadanie 1: Numeryczne Dopasowanie Parametrów Rozkładów Prawdopodobieństwa**

**Cel:** Celem tego zadania jest praktyczne przećwiczenie metod estymacji parametrów dla wybranych rozkładów prawdopodobieństwa. Dla każdego z poniższych rozkładów należy:

1. Wygenerować dane syntetyczne (próbkę losową) o z góry określonych parametrach.  
2. Następnie, korzystając z metod numerycznych (np. metoda największej wiarygodności), dopasować parametry teoretycznego rozkładu do wygenerowanych danych.  
3. Porównać otrzymane estymaty z rzeczywistymi wartościami parametrów użytymi do generacji danych.

**Rozkłady do analizy:**  
a. Rozkład Bernoulliego: Estymuj parametr p (prawdopodobieństwo sukcesu).  
b. Rozkład Poissona: Estymuj parametr λ (średnia liczba zdarzeń w jednostce czasu/przestrzeni).  
c. Rozkład Normalny (Gaussa): Estymuj parametry μ (wartość oczekiwana) oraz σ (odchylenie standardowe).

### **Zadanie 2: Budowa i Analiza Bayesowskiego Modelu Liniowego**

**Cel:** W tym zadaniu skupimy się na implementacji, dopasowaniu oraz dogłębnej analizie bayesowskiego modelu regresji liniowej.

1. **Implementacja i Dopasowanie Modelu:**  
   * Zaimplementuj od podstaw bayesowski model liniowy.  
   * Wygeneruj syntetyczny zbiór danych, dla którego znasz rzeczywiste parametry modelu liniowego (nachylenie i wyraz wolny).  
   * Dopasuj swój bayesowski model liniowy do tych syntetycznych danych, uzyskując rozkłady a posteriori dla parametrów modelu.  
2. **Badanie Wpływu Szumu i Wielkości Zbioru Danych:**  
   * Przeprowadź serię eksperymentów, aby zbadać, jak zmieniają się wyniki dopasowania modelu (w szczególności rozkłady a posteriori parametrów) w zależności od:  
     * Poziomu szumu w danych, reprezentowanego przez odchylenie standardowe σ:  
       * σ=0.05  
       * σ=0.1  
       * σ=0.5  
     * Liczby obserwacji (punktów danych) N:  
       * N=3  
       * N=10  
       * N=1000  
   * W swojej analizie odnieś się do dwóch rodzajów niepewności:  
     * **Niepewność aleatoryczna:** Związana z inherentnym, losowym szumem w danych. W modelu jest ona często reprezentowana przez parametr σ.  
     * **Niepewność epistemiczna:** Wynikająca z braku wystarczającej ilości danych do jednoznacznego określenia parametrów modelu. Odzwierciedla się ona w szerokości rozkładów a posteriori parametrów.  
3. **Testowanie Modelu na Danych Nieliniowych:**  
   * Wygeneruj dane z modelu nieliniowego, np. funkcji kwadratowej: y=0.6x2+0.5x−0.3.  
   * Spróbuj dopasować swój bayesowski model *liniowy* do tych nieliniowych danych.  
   * Oceń jakość dopasowania i skomentuj, jak model liniowy radzi sobie z próbą aproksymacji zależności nieliniowej. Jakie wnioski można wyciągnąć na temat ograniczeń modelu?

3. **Analiza Wyników:**

````python
import numpy as np
import tensorflow_probability as tfp
import tensorflow as tf

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import collections
tfd=tfp.distributions
tfb=tfp.bijectors
sns.set_theme('notebook')
````

generujemy dane

````python
np.random.seed(44)
x = np.random.normal(0., 1.,size=40)
x = np.sort(x)
scale=0.5
y = 0.9*x - 0.5 + np.random.normal(scale=scale, size=x.shape)

#y = 2*x + 0.9* x**2+4 + np.random.normal(scale=0.4, size=x.shape)

# zły model +0.6*x**2

# x=np.random.rand(100)*2-1
# scale=0.2
#y = 0.5*x  - 0.3 + np.random.normal(scale=scale, size=x.shape)
````

````python
sns.scatterplot(x=x,y=y)
````

````python
joint = tfd.JointDistributionNamed(dict(
    beta0 = tfd.Laplace(loc=0., scale=2.),
    beta1=tfd.Normal(loc=0, scale=2.),
    sigma=tfd.Gamma(4., 4.),
    y=lambda beta0, beta1, sigma: tfd.Independent(tfd.Normal(
        loc = beta0[..., tf.newaxis] + beta1[..., tf.newaxis]*x[np.newaxis, ...],
        scale = sigma[..., tf.newaxis]
    ), reinterpreted_batch_ndims=1)
))
````

````python
joint2 = tfd.JointDistributionNamedAutoBatched(dict(
    beta0 = tfd.Laplace(loc=0., scale=2.),
    beta1=tfd.Normal(loc=0, scale=2.),
    sigma=tfd.Gamma(4., 4.),
    y=lambda beta0, beta1, sigma: tfd.Independent(tfd.Normal(
        loc = beta0 + beta1*x,
        scale = sigma
    ), reinterpreted_batch_ndims=1)
))

#s2=joint2.sample(2)

#joint2.log_prob(s2)
````

````python
joint3 = tfd.JointDistributionNamedAutoBatched(dict(
    beta0 = tfd.Laplace(loc=0., scale=2.),
    beta1=tfd.Normal(loc=0, scale=2.),
    beta2=tfd.Normal(loc=0, scale=2.),
    sigma=tfd.Gamma(4., 4.),
    y=lambda beta0, beta1,beta2, sigma: tfd.Independent(tfd.Normal(
        loc = beta0 + beta1*x + beta2*x**2,
        scale = sigma
    ), reinterpreted_batch_ndims=1)
))

````

````python

def target_log_prob(beta0, beta1, sigma):
  return joint.log_prob(beta0=beta0,beta1=beta1, sigma=sigma, y=y[np.newaxis,...])


y_dist = tfp.distributions.Normal(
  loc=tfp.util.DeferredTensor(tf.Variable([0.0, 1.0, 1.0]),
                              lambda beta: beta[0] + beta[1]*x+beta[2]*x**2,
                              shape=x.shape),
  scale=tfp.util.TransformedVariable([1.], bijector=tfp.bijectors.Softplus())
  )
losses = tfp.math.minimize(lambda : -tf.reduce_sum(y_dist.log_prob(y)),
                           num_steps=2000,
                           optimizer=tf.optimizers.Adam(learning_rate=0.2))
````

## MCMC

````python
start_dist = tfd.JointDistributionNamed(dict(
    beta0=tfd.Normal(y_dist.variables[0][0],0.01),
    beta1=tfd.Normal(y_dist.variables[0][1],0.02),
    # beta2=tfd.Normal(y_dist.variables[0][1],0.02),
    sigma=tfd.Normal(y_dist.stddev()[0],0.006)
))
````

````python

hmc = tfp.mcmc.NoUTurnSampler(
    target_log_prob_fn=target_log_prob,
    step_size=[.015,.025, 0.0065])

hmc = tfp.mcmc.TransformedTransitionKernel(
    inner_kernel=hmc,
    bijector=[
              tfp.bijectors.Identity(),
              tfp.bijectors.Identity(),
            #   tfp.bijectors.Identity(),
              tfp.bijectors.Softplus(),  # s1
    ])

hmc = tfp.mcmc.DualAveragingStepSizeAdaptation(
    inner_kernel=hmc,
    num_adaptation_steps=1000,
    target_accept_prob=.7)

joint_sample = joint2.sample(10)
joint_sample = start_dist.sample(5)
#joint_sample = q.sample(5)
initial_state = [
    joint_sample['beta0'],
    joint_sample['beta1'],
    # joint_sample['beta2'],
    joint_sample['sigma'],
]

@tf.function(autograph=False, jit_compile=True)
def run():
  return tfp.mcmc.sample_chain(
      num_results=500,
      current_state=initial_state,
      kernel=hmc,
      num_burnin_steps=100,
      trace_fn=lambda _, kr: kr)

samples, traces = run()
for sample in samples:
  print('R-hat diagnostics: ', tfp.mcmc.potential_scale_reduction(sample))
````

## SVI

````python
q = tfd.JointDistributionNamed(dict(
                                     beta0=tfd.Normal(loc=tf.Variable(0.),
                                                scale=tfp.util.TransformedVariable(1.5, tfp.bijectors.Softplus(),
                                                    )),
                                      beta1=tfd.Normal(loc=tf.Variable(0.),
                                                scale=tfp.util.TransformedVariable(1.5, tfp.bijectors.Softplus(),
                                                    )),
                                      sigma=tfd.Normal(loc=tf.Variable(2.),
                                                scale=tfp.util.TransformedVariable(0.01, tfp.bijectors.Softplus(),
                                                    ))

))

````

````python
losses = tfp.vi.fit_surrogate_posterior(
    target_log_prob,
    surrogate_posterior=q,
    optimizer=tf.optimizers.Adam(learning_rate=0.1),
    num_steps=1000,
    sample_size=10)
````

````python
print(l3),1/a3.mean()
````

````python
def l3_fn(l1,l2,p):
  return tfd.Deterministic((1.-p)*(l1+l2))

model = tfd.JointDistributionNamed(dict(
    l1 = tfd.Gamma(2.,2/100.),
    l2 = tfd.Gamma(2.,2/100.),
    p = tfd.Beta(1.,1.),
    l3=l3_fn,
    a1 = lambda l1: tfd.Sample(tfd.Exponential(l1),len(a1)),
    a2 = lambda l2: tfd.Sample(tfd.Exponential(l2), len(a2)),
    a3 = lambda l3: tfd.Sample(tfd.Exponential(l3), len(a3)),
))
````

````python
q=tfd.JointDistributionNamed(dict(
    l1 = tfd.TransformedDistribution(tfd.Normal(tf.Variable(1./a1.mean(), dtype=tf.float32),
                    tfp.util.TransformedVariable(1.,tfp.bijectors.Softplus(), dtype=tf.float32)
                    ),tfb.Softplus()),
    l2 = tfd.TransformedDistribution(tfd.Normal(tf.Variable(np.float32(1.)/a2.mean()),
                    tfp.util.TransformedVariable(1.,tfp.bijectors.Softplus(), dtype=tf.float32)
                    ),tfb.Softplus()),
    p = tfd.TransformedDistribution(tfd.Normal(tf.Variable(np.float32(1.)-(np.float32(1.)/a1.mean())/(np.float32(1.)/a1.mean()+np.float32(1.)/a2.mean())),
                   tfp.util.TransformedVariable(1.,tfp.bijectors.Softplus(), dtype=tf.float32)
                   ),tfb.Sigmoid()),
    l3 = l3_fn

))
````

````python
losses = tfp.vi.fit_surrogate_posterior(
    lambda l1,l2,l3, p: model.log_prob(l1=l1,l2=l2,l3=l3,p=p,a1=a1, a2=a2,a3=a3),
    surrogate_posterior=q,
    optimizer=tf.optimizers.Adam(learning_rate=0.1),
    num_steps=1000,
    sample_size=50)
````
